## Korrelationen

Korrelationen sind ein grundlegendes Werkzeug in der Statistik, um den Zusammenhang zwischen zwei Variablen zu untersuchen. Sie geben an, ob und wie stark zwei Gr√∂√üen miteinander in Beziehung stehen.

Die Korrelation wird durch den Korrelationskoeffizienten beschrieben, der Werte zwischen -1 und +1 annehmen kann:

- +1: Perfekte **positive Korrelation** (wenn eine Variable steigt, steigt die andere auch proportional).
- 0: **Keine Korrelation** (kein erkennbarer Zusammenhang).
- -1: Perfekte **negative Korrelation** (wenn eine Variable steigt, sinkt die andere proportional).

In der Praxis verwenden wir h√§ufig den **Pearson-Korrelationskoeffizienten**, um lineare Zusammenh√§nge zu messen. Voraussetzung daf√ºr ist, dass wir zwei numerische Variablen haben.

F√ºr andere Situationen gibt es andere Korrelations-Tests. Diese sind aber nicht so h√§ufig und werden in diesem Kurs nicht betrachtet.

In Python gibt es verschiedene M√∂glichkeiten, Korrelation zu berechnen:

Hinweis zur Interpretation der `GPA`: Im Gegensatz zum deutschen Schulsystem entspricht eine *h√∂here* `GPA` einer *besseren* Leistung.  

```python
# Mit pandas w√§hlen wir eine Spalte aus
# Dann bestimmen wir die Funktion die verwendet werden soll (corr())
# Und w√§hlen als Argument f√ºr die corr() Funktion die zweite Spalte aus
df.Study_Hours_Per_Day.corr(df.GPA)
```

## Ausrei√üeranalyse
Ausrei√üer sind solche Datenpunkte, die von der Verteilung der restlichen Datenpunkte stark abweichen. Diese k√∂nnen das Ergebnis verzerren und auf einen Messfehler hinweisen.

Beobachten wir zum Beispiel Datenpunkte die 22 Stunden am Tag lernen, dann ist das vermutlich ein Fehler in der Datenerhebnung und sollte entfernt werden. Aber Achtung: Nur weil ein Datenpunkt ungew√∂hnlich ist, sollten wir ihn nicht unbedingt entfernen. Ungew√∂hnliche Dinge passieren, und wenn wir diese nachtr√§glich entfernen (oder anderweitig korrigieren), sind wir vielleicht diejenigen die das Ergebnis verzerren und falsche R√ºckschl√ºsse ziehen.

Daher die Faustregel: Ausrei√üer sollten untersucht werden aber Korrekturen sollten wir nur vornehmen, wenn wir
a) einen Grund haben zu glauben, dass ein Messfehler vorliegt
b) unsere statistischen Aussagen nur auf "Mehrheit" einschr√§nken wollen und die extrema bewusst au√üer Acht lassen wollen.

Wie finden wir also heraus ob ein Datenpunkt ein Ausrei√üer ist?

## Normalverteilung
Statistische Tests erfodern oftmals, dass die Variablen die getestet werden sollen, normalverteilt sind. Diese Verteilung ist weit Verbreitet, auch unter dem Namen Gauss Verteilung oder Glockenkurve bekannt.

Wie pr√ºfen wir, ob Variablen normalverteilt sind?

### Visuelle Inspektion
Ein Histogramm gibt uns ein erstes Indiz daf√ºr ob eine Variable normalverteilt ist. In diesem Fall sehen wir die klassische Glockenform. Auf den ersten Blick wirkt die Variable `GPA` also normalverteilt.

```python
df.GPA.hist()
```

### Empirischer Test
Normalit√§t kann man auch empirisch testen, geeignet sind daf√ºr der **Shapiro-Wilk Test** und der **Kolgomorov-Smirnov (KS) Test**, benannt nach den Erfindern.


**Shapiro Wilk** ist besonders f√ºr kleine Stichproben geeignet (<2000 Datenpunkte, aber auch <50)

**KS** erforder gr√∂√üere Datenmengen von mindestens 50, um aussagekr√§ftig zu sein.

Lasst uns erstmal z√§hlen wie viele Datenpunkte wir haben

```python
len(df)
```

Wir haben exakt 2000 Datenpunkte in unserem Sample. Daher ist der **KS-Test f√ºr Normalit√§t** vermutlich der angemessenere. Aber ich zeige euch trotzdem beide:

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test
stat, p_value = shapiro(df.GPA)

print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"P-Value: {p_value}")

# Interpret result
if p_value > 0.05:
    print("Data is likely normal")
else:
    print("Data is not normal")
```

```python
from scipy.stats import kstest

# Perform Kolmogorov-Smirnov test for normality
stat, p_value = kstest(df.GPA, 'norm')

print(f"Kolmogorov-Smirnov Test Statistic: {stat}")
print(f"P-Value: {p_value}")

# Interpret result
if p_value > 0.05:
    print("Data is likely normal")
else:
    print("Data is not normal (reject H0)")
```

Beide Tests sagen uns, dass die Verteilung nicht normal ist. Noch deutlicher ist das Ergebnis, wenn wir stattdessen die Variable `Study_Hours` pr√ºfen.

Somit sollten wir mit der Anwendung und Interpretation der **Pearson Korrelation** vorsichtig sein.
